```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
dat = pd.read_csv('cancer_data.csv')
```

```python
dat.head()
```

```python
dat.drop('Unnamed: 32',axis=1)
```

```python
diagnosis_to_num = {'B': 0,
                    'M': 1}
dat['diagnosis'] = dat['diagnosis'].map(diagnosis_to_num)
```

```python
y = dat[['diagnosis']]
```

```python
X = dat
X.pop('diagnosis')
X.pop('Unnamed: 32')
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_clf = DecisionTreeClassifier(random_state=42)
```

```python
tree_clf.fit(X_train,y_train)
```
