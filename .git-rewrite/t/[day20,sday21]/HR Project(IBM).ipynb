# HR Project 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import tree
```

```python
df = pd.read_csv('hrdata.csv')
df.pop("EmployeeCount")
df.pop("EmployeeNumber")
df.pop("StandardHours")
df.pop("Over18")
```

```python
X = df
y = df['Attrition']
```

```python
X.pop('Attrition')
```

```python
y.unique()
```

```python
from sklearn import preprocessing
ban = preprocessing.LabelBinarizer()
```

```python
y = ban.fit_transform(y)
```

```python
df.select_dtypes(['object'])
```

```python
ind_BusinessTravel = pd.get_dummies(df['BusinessTravel'],prefix='BusinessTravel')
ind_Department = pd.get_dummies(df['Department'],prefix='Department')
ind_EducationField = pd.get_dummies(df['EducationField'],prefix='EducationField ')
ind_Gender = pd.get_dummies(df['Gender'],prefix='Gender')
ind_MaritalStatus = pd.get_dummies(df['MaritalStatus'],prefix='MaritalStatus')
ind_JobRole = pd.get_dummies(df['JobRole'],prefix='JobRole')
ind_OverTime = pd.get_dummies(df['OverTime'],prefix='OverTime')
```

```python
df_fin = pd.concat([ind_BusinessTravel,ind_Department,ind_EducationField,ind_Gender,ind_JobRole,ind_MaritalStatus,ind_OverTime,df.select_dtypes(['int64'])],axis=1)
```

```python
df_fin.shape
```

# Decision tree

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_fin, y)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf  = DecisionTreeClassifier(random_state=42)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
